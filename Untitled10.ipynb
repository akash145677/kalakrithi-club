```python
from google.colab import files

# Upload your two text files
print("Upload exactly TWO text files to check plagiarism:")
uploaded = files.upload()

# Extract the uploaded filenames
files_uploaded = list(uploaded.keys())

# Validate file count
if len(files_uploaded) != 2:
    print("‚ùó Error: Please upload exactly TWO .txt files.")
else:
    file1, file2 = files_uploaded[0], files_uploaded[1]

    # Read content
    text1 = open(file1, "r", encoding="utf-8").read()
    text2 = open(file2, "r", encoding="utf-8").read()

    # Functions defined earlier
    import math, string

    def clean_text(text):
        text = text.lower()
        return text.translate(str.maketrans('', '', string.punctuation))

    def word_freq(text):
        words = clean_text(text).split()
        freq = {}
        for w in words:
            freq[w] = freq.get(w, 0) + 1
        return freq

    def cosine_similarity(freq1, freq2):
        all_words = set(freq1.keys()).union(set(freq2.keys()))
        dot = sum(freq1.get(w, 0) * freq2.get(w, 0) for w in all_words)
        mag1 = math.sqrt(sum(freq1.get(w, 0)**2 for w in all_words))
        mag2 = math.sqrt(sum(freq2.get(w, 0)**2 for w in all_words))
        if mag1 == 0 or mag2 == 0:
            return 0
        return (dot / (mag1 * mag2)) * 100

    # Calculate plagiarism
    freq1 = word_freq(text1)
    freq2 = word_freq(text2)

    plagiarism_percent = round(cosine_similarity(freq1, freq2), 2)

    print("\nüìÑ File 1:", file1)
    print("üìÑ File 2:", file2)
    print("\nüî• PLAGIARISM DETECTED:", plagiarism_percent, "%")
```
